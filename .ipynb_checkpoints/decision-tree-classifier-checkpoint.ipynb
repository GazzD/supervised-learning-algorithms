```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
# Cargamos el conjunto de datos de iris
iris = load_iris()
print(iris.feature_names)
```

```python
print(iris.target_names)

# Para tener los datos de verificación eliminamos unas etiquetas
# El elemento 0 pertenece a la especie Setosa
# El elemento 50 pertenece a la especie Versicolor
# El elemento 100 pertenece a la especie Virginica
removed = [0, 50, 100]
new_target = np.delete(iris.target, removed)
new_data = np.delete(iris.data, removed, axis=0)

# Usamos el clasificador de árbol de decisión de scikit-learn para la clasificación
# Entrenar clasificador
clf = tree.DecisionTreeClassifier() # definimos árbol de decisión clasificador
clf = clf.fit(new_data, new_target) # entrenamos datos sobre new_data y new_target
prediction = clf.predict(iris.data[removed]) # asignar los datos eliminados a la entrada

# Comprobamos si nuestras etiquetas predichas coinciden con las etiquetas originales
print("Original Labels", iris.target[removed])
print("Labels predicted", prediction)

# Dibujamos el árbol
tree.plot_tree(clf)
```
